```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update(**{'figure.dpi':150})
```

```python
data = pd.read_csv('On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv')
```

```python
# 1 set of variables: airline vs cancellation rate
airlines = list(set(data['Reporting_Airline']))
cancel_counts = []
for j in range(len(airlines)):
    sub_array = data[data['Reporting_Airline'].str.contains(airlines[j])]
    flights = sub_array.shape[0]
    cancel_array = sub_array['Cancelled']
    total_cancellation = sum(cancel_array)
    cancel_counts.append(total_cancellation/flights)
plt.bar(airlines, cancel_counts)
plt.xlabel('airline')
plt.ylabel('cancellation rate')
plt.title('Comparison of cancellation rates between airlines')
plt.show()
```

From the graph above, we can see that the two airlines with a relatively high cancellation rate of around 7% are the airlines MQ and EV. The HA airline has a very low cancellation rate among the airlines.

```python
# 2nd set of variables: weekday and flight speed
filtered = data[data['Cancelled'] == 0]
filtered['flight_speed'] = filtered.apply(lambda filtered: filtered['Distance']/filtered['ActualElapsedTime'], axis = 1)
stat = filtered['flight_speed'].describe()
filtered['standard_speed'] = filtered.apply(lambda filtered: (filtered['flight_speed'] - stat['mean'])/stat['std'], axis = 1)
```

```python
speed = []
for i in range(1,8):
    sub_array = filtered[filtered['DayOfWeek'] == i]
    flights = sub_array.shape[0]
    speed_array = sub_array['standard_speed']
    total_speed = speed_array.sum()
    speed.append(total_speed/flights)

plt.bar(['Monday', 'Tuesday', 'Wed', 'Thursday', 'Friday', 'Sat', 'Sun'], speed)
plt.xlabel('day of the week')
plt.ylabel('S-score standardized averaged speed')
plt.title('Comparison between day of the week and standardized average speed')
plt.show()
```

From the data, it seems like the average flight speed on weekends is greater than the average flight speed on weekdays, with Saturday being the day with the greatest normalized average flight speed among all the days of the week.
